## ***Observations from Visualizations***

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True )
plt.show()
```

***Maximum are from Binders, Paper, furnishings, Phones, storage, art, accessories and minimum from copiers, machines, suppliers***

```python
#valuecounts for State
df['State'].value_counts()
```

```python
plt.figure(figsize=(20,12))
sns.countplot(x='State',data=df,palette='rocket',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

***Highest number of buyers are from California, New York and Texas***

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

***Most customers tends to buy quantity of 2 and 3 and Maximum Discount given is 0 to 20 percent***

```python
plt.figure(figsize=(10,10))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

***No correlation between profit and discount***

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','brown'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

***Profit and sales are maximum in consumer segment and minimum in Home Office segment***

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='PuBuGn')
plt.xlabel('Region',fontsize=15)
plt.show()
```

***Segment wise sales are almost same in every region***

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['darkblue','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

***Profit and sales are maximum in west region and minimum in south region***

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','blue'],figsize=(20,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

***High profit is for California, New york and loss is for Texas, Pennsylvania and Ohio***

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['black','gray'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```
